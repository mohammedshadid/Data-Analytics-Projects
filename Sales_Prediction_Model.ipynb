# Sales Prediction Model using Machine Learning

## Problem Statement
Predict sales based on profit, quantity, and discount.

## Tools Used
- Python
- Pandas
- Scikit-learn

## Approach
- Data cleaning
- Model building (Linear Regression & Random Forest)
- Model evaluation using R2 score

## Results
- Linear Regression Accuracy: ~0.37
- Random Forest Accuracy: ~0.78

## Business Impact
Helps businesses predict future sales and make better decisions.

```python
import pandas as pd
```

```python
df = pd.read_excel("/content/Superstore.xlsx")
df.head()
```

```python
df = df[['Sales', 'Profit', 'Quantity', 'Discount']]
df.head()
```

```python
X = df[['Profit', 'Quantity', 'Discount']]
y = df['Sales']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
predictions[:5]
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, predictions)
```

```python
from sklearn.ensemble import RandomForestRegressor

model2 = RandomForestRegressor()
model2.fit(X_train, y_train)

pred2 = model2.predict(X_test)

from sklearn.metrics import r2_score
r2_score(y_test, pred2)
```
