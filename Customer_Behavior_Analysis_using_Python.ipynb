# Customer Behavior Analysis using Python

## Objective
Analyze sales data to understand:
- Sales trends over time
- Performance by category
- Top customers
- Regional insights

## Dataset
Superstore dataset

```python
import pandas as pd
from google.colab import files
import io

uploaded = files.upload()

# Assuming the user uploads 'Superstore.xlsx' based on the original code
if 'Superstore.xlsx' in uploaded:
    df = pd.read_excel(io.BytesIO(uploaded['Superstore.xlsx']))
    print("Superstore.xlsx loaded successfully.")
    df.head()
else:
    print("Error: 'Superstore.xlsx' not found among uploaded files. Please upload the correct file.")
```

```python
top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)
top_customers
```

```python
import matplotlib.pyplot as plt

top_customers.plot(kind="bar")
plt.title("Top 10 Customers by Sales")
plt.show()
```

## Insight

A small number of customers contribute significantly to total sales.
These customers are highly valuable and should be prioritized for retention.

```python
df.groupby("Segment")["Sales"].sum()
```

## Insight

Consumer segment generates the highest sales compared to Corporate and Home Office.
This indicates strong demand from individual customers.

```python
df.groupby("Region")["Sales"].sum()
```

## Insight

West and East regions generate higher sales.
Other regions show lower performance and can be improved.

```python
df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()

monthly_sales.plot()
```

## Sales Trend Analysis

```python
import matplotlib.pyplot as plt

monthly_sales.plot(figsize=(10,5))

plt.title("Monthly Sales Trend", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Sales")

plt.grid()
plt.show()
```

## Insight

Sales show an overall increasing trend over time.
There are fluctuations indicating seasonal demand patterns.

## Sales By Category

```python
df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Category")
plt.show()
```

## Insight

Technology category generates the highest sales.
Furniture and Office Supplies perform slightly lower.

## Profit Analysis

```python
df.groupby("Category")["Profit"].sum()
```

## Insight

Some categories generate higher profits than others.
This indicates that revenue does not always translate into profit, and cost management is important.
